Playing with Graphs

```python
from rdflib import Graph
```

```python
g = Graph()
```

```python
#parsing a RDF graf from dbpedia
g.parse('https://dbpedia.org/resource/Ludwig_Wittgenstein')
```

```python
#iterate through the grap triple: subject, predicate, object.
for i, (sub,pred,obj) in enumerate(g):
  print(sub,pred,obj)
  if i == 5:
    break
```

```python
#printing only the subject
for i, (sub,pred,obj) in enumerate(g):
  print(sub)
  if i == 5:
    break
```

```python
#show the graph size
print(f'Size: {len(g)} facts')
```

```python
#print the graph in RDF turtle format
print(g.serialize(format='ttl'))
```
